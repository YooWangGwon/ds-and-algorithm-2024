1번 문제

```python
## 함수 선언
def isStFull():
    global SIZE, stack, top
    if top >= (SIZE-1):
        return True
    else:
        return False

def isStEmpty():
    global SIZE, stack, top
    if top == -1:
        return True
    else:
        return False
    
def push(data):
    global SIZE, stack, top
    if isStFull() == True:
        print('스택이 꽉 찼습니다.')
        return
    top += 1
    stack[top] = data

def pop():
    global SIZE, stack, top
    if isStEmpty() == True:
        print('스택이 비어있습니다')
        return None
    else:
        data = stack[top]
        stack[top] = None
        top -= 1
        return data

def peek():
    global SIZE, stack, top
    if isStEmpty() == True:
        print('스택이 비어있습니다')
        return None
    else:
        return stack[top]

# 전역변수 선언
SIZE = int(input('스택의 `크기를 입력하세요. > '))
stack = [None for _ in range(SIZE)]
top = -1

# 메인코드
if __name__ == '__main__':

    while True:
        select = input('PUSH(p),POP(o),PEEK(e),Exit(x) 중 하나를 선택하세요. > ')

        if select.lower() == 'p':
            data = input('입력할 데이터 -> ')
            push(data)
            print(f'스택 상태 : {stack}')
        elif select.lower() == 'o':
            data = pop()
            print(f'추출된 데이터 : {data}')
            print(f'스택 상태 : {stack}')
        elif select.lower() == 'e':
            data = peek()
            print(f'확인된 데이터 : {data}')
            print(f'스택 상태 : {stack}')
        elif select.lower() == 'x':
            break
        else:
            continue
        
print('프로그램 종료')
```

2번 문제

```python
# 함수 선언

def isQueueFull():
    global SIZE, queue, front, rear
    if rear != (SIZE-1):
        return False
    elif rear == (SIZE-1) and front == -1:
        return True
    else:
        while front != -1:
            for i in range(front+1, SIZE):
                queue[i-1] = queue[i]
                queue[i]
            front -= 1
            rear -= 1
        return False

def isQueueEmpty():
    global SIZE, queue, front, rear
    if front == rear:
        return True
    else:
        return False

def enQueue(data):
    global SIZE, queue, front, rear
    if isQueueFull():
        print('대기줄이 꽉 찼습니다.')
        return
    else:
        rear += 1
        queue[rear] = data

def deQueue():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('대기줄이 비어있습니다')
        return None

    front += 1
    data = queue[front]
    queue[front] = None

    for i in range(front+1, SIZE):
        queue[i-1] = queue[i]
        queue[i] = None
    front -= 1
    rear -= 1

    return data
    
def peek2():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('대기줄이 비어있습니다')
        return None
    return queue[front+1]

# 전역변수
SIZE = int(input('대기줄 크기 입력(정수) > '))
queue = [None for _ in range(SIZE)]
front = rear = -1

# 메인코드
if __name__ == '__main__':
    while True:
        select = input('대기줄에 손님 추가(e), 가게에 손님 입장(d), 다음 손님 확인(p), 종료(x) > ')

        if select.lower() == 'e':
            data = input('대기줄에 추가되신 손님 성함 입력 -> ')
            enQueue(data)
            print(f'현재 대기줄 상태 : {queue}')
        elif select.lower() == 'd':
            data = deQueue()
            print(f'들어가신 손님 : {data}')
            print(f'현재 대기줄 상태 : {queue}')
        elif select.lower() == 'p':
            data = peek2()
            print(f'다음에 들어가실 손님 : {data}')
            print(f'현재 대기줄 상태 : {queue}')
        elif select.lower() == 'x':
            break

        else:
            continue

print('프로그램 종료')  
```

3번 문제

```python
import random

# 클래스 선언
class TreeNode():
    left = right = data = None

    def __init__(self) -> None:
        self.left = self.right = self.data = None

# 전위 순회 함수
def preorder(node):
    if node == None: return

    print(node.data, end=' ')
    preorder(node.left)
    preorder(node.right)

# 전역 변수
root = None
itemAry = ['coffee','milk','coke','candy','water','snack']
sales = int(input('판매량을 입력하세요(정수만) > '))
sellAry = [random.choice(itemAry) for _ in range(sales)]

print(f'오늘 판매된 물건(중복 O) -> {sellAry}')


# 메인코드
node = TreeNode()
node.data = sellAry[0]
root = node

for name in sellAry[1:]:
    node = TreeNode()
    node.data = name

    curr = root
    while True:
        if name == curr.data:
            break
        if name < curr.data: 
            if curr.left == None: 
                curr.left = node
                break 
            else:
                curr = curr.left
        else:
            if curr.right == None:
                curr.right = node
                break
            else:
                curr = curr.right

print('이진트리 탐색 구성 완료')
print('오늘 판매된 종류(중복 X) -> ',end = ' ')
preorder(root)
```
